```python
datapath = '/ptmp/wangqi/transfer_folder/LS200X_Norm/crops'
device = torch.device('cuda:0')
batch_size = 4
lr = .0001
data = DataLoader(CropDataset(datapath),batch_size = batch_size,drop_last=True,num_workers=8,shuffle=True)
model_name = "c15"
#Tensor = torch.cuda.HalfTensor # half precision is used
Tensor = torch.cuda.FloatTensor
```

```python
def weights_init(m):
    if isinstance(m, nn.Conv3d):
        torch.nn.init.xavier_uniform_(m.weight)
        torch.nn.init.zeros_(m.bias)
    if isinstance(m,nn.BatchNorm3d):
        torch.nn.init.normal_(m.weight, 0.0, 0.02)
        torch.nn.init.constant_(m.bias, 0)
```

```python
criterion_GAN = torch.nn.BCEWithLogitsLoss().to(device)
criterion_content = torch.nn.L1Loss().to(device)
criterion_pixel = torch.nn.L1Loss().to(device)

checkpoint = 0
save_every_iter = 300

Dnet = Discriminator().to(device)
Gnet = Generator().to(device)
#Gnet = Generator()
#FE = FeatureExtractor().to(device)
FE = resnet10( sample_input_W=32,
                sample_input_H=32,
                sample_input_D=32,num_seg_classes=32).to("cuda:0") # pre-trained FE
freeze_model(FE)
    
```

```python
optimizer_G = torch.optim.Adam(Gnet.parameters(), lr=lr, betas=(0.9, .999))
optimizer_D = torch.optim.Adam(Dnet.parameters(), lr=lr, betas=(0.9, .999))
if not checkpoint == 0:
    ckp = torch.load("../saved_models/%s_Crop_%d_%d.pth"%(model_name,checkpoint,2*save_every_iter-1))
    #checkpoint_model_G = f'../saved_models/{model_name}_generator_{checkpoint}_{len(data)-1}.pth'
    #checkpoint_model_D = f'../saved_models/{model_name}_discriminator_{checkpoint}_{len(data)-1}.pth'
    Dnet.load_state_dict(ckp['Dnet_state_dict'])
    Gnet.load_state_dict(ckp['Gnet_state_dict'])
    FE.load_state_dict(ckp['FE_state_dict'])
    optimizer_D.load_state_dict(ckp['optimizer_D_state_dict'])
    optimizer_G.load_state_dict(ckp['optimizer_G_state_dict'])
    
```

```python
def plot_val(sr,hr,title):
    fig,ax = plt.subplots(2,4,figsize=(20,10))
    axe = ax.ravel()
    fig.suptitle(title)
    with torch.no_grad():
        psnr_v = psnr(sr.squeeze().cpu().numpy(),hr.squeeze().numpy())
        ssim_v = ssim(sr.squeeze().cpu().numpy(),hr.squeeze().numpy())
    axe[0].text(5,5,f"psnr{psnr_v:.3f}",fontdict={'color':'red','size':16,'weight':'bold'})
    axe[0].text(5,9,f"SSIM{ssim_v:.3f}",fontdict={'color':'red','size':16,'weight':'bold'})
    for i in range(hr.size()[0]):
        axe[i].imshow(hr[i,0,32].squeeze().detach().numpy(),cmap='gray')
        axe[i+4].imshow(sr[i,0,32].squeeze().detach().cpu().numpy(),cmap='gray')
    plt.show()
```

## Loss function

$Loss_{Gnet} = L_{1}(FE(\hat{x}),FE(x)) + 5*10^{-3}*BCE(D(\hat{x}) - \bar{D(x)},Real) + 1*10^{-2}*L_{1}(\hat{x},x)$

```python

warm_up = True # whether want a warm up session for Gnet

#-----------writing log--------------
writer = SummaryWriter(log_dir=f'{model_name}/run1',comment="Jupyter_ESRGAN19_5G1D")
hparams={'lr': lr, 'bsize': batch_size, 'Gnet_warmup':warm_up}
metrics={'Loss/train_D_loss':None,
         'Loss/train_G_loss':None,
         'Loss/train_GAN_loss':None,
         'Loss/train_pixel_loss':None,
         'Loss/train_content_loss':None,
         'Loss/train_psnr':None,
         'Loss/train_ssim':None
        }

dataloader = data
#[m.apply(weights_init) for m in [Dnet,Gnet,FE]]
valid = Variable(Tensor(np.ones((batch_size, *Dnet.output_shape))), requires_grad=False)
fake = Variable(Tensor(np.zeros((batch_size, *Dnet.output_shape))), requires_grad=False)
for epoch in range(num_epoch):
    for i,imgs in enumerate(dataloader):
        imgs_hr = imgs['hr'].to(device)
        imgs_lr = imgs['lr'].to(device)
        
        optimizer_G.zero_grad() # clean all grad at begining of update

    
        gen_hr = Gnet(imgs_lr) 
        loss_pixel = criterion_pixel(gen_hr,imgs_hr)

        if (epoch==0 and warm_up == True):
            loss_pixel.backward()
            optimizer_G.step()
            print(
                "[Epoch %d/%d] [Batch %d/%d] [G pixel: %f]"
                % (epoch, num_epoch, i, len(dataloader), loss_pixel.item())
                 )   
            #print(f"device 0 : {(torch.cuda.memory_reserved(0) - torch.cuda.memory_allocated(0))/(1024**2) }MB \n")
            continue

        pred_real = Dnet(imgs_hr).detach()
        pred_fake = Dnet(gen_hr)
        loss_GAN = criterion_GAN(pred_fake - pred_real.mean(0, keepdim=True), valid)
        #print(f"device 0 : {(torch.cuda.memory_reserved(0) - torch.cuda.memory_allocated(0))/(1024**2) }MB \n")

        gen_features = FE(gen_hr) # [ ] memo overflow
        real_features = FE(imgs_hr).detach()

        loss_content = criterion_content(gen_features, real_features)
        loss_G = loss_content + 5e-3 * loss_GAN + 1e-2 * loss_pixel

        loss_G.backward()
        optimizer_G.step()
        
        
        #################
        # Discriminator
        #################
        optimizer_D.zero_grad()
        
        pred_real = Dnet(imgs_hr)
        pred_fake = Dnet(gen_hr.detach())

        # Adversarial loss for real and fake images (relativistic average GAN)
        loss_real = criterion_GAN(pred_real - pred_fake.mean(0, keepdim=True), valid)
        loss_fake = criterion_GAN(pred_fake - pred_real.mean(0, keepdim=True), fake)

        # Total loss
        loss_D = (loss_real + loss_fake) / 2

        if i%5 == 0: # update D net every 5 Gnet step
            loss_D.backward()
            optimizer_D.step()
        

        ##################
        # scheduler update
        ##################
        #scheduler_D.step()
        #scheduler_G.step()
        #[print(sched.get_last_lr()) for sched in [scheduler_D,scheduler_G]] # add print lr per iteration

        
        # Write Gnet gradient at end of epoch
        '''
        if not i%5:    
            for tag, parm in Gnet.named_parameters():
                writer.add_histogram("Gnet/"+tag, parm.grad.data.cpu().numpy(), epoch)
            for tag, parm in Dnet.named_parameters():
                writer.add_histogram("Dnet/"+tag, parm.grad.data.cpu().numpy(), epoch)
        '''
        
        with torch.no_grad(): 

            # batch-wise PSNR, SSIM calculation
            
            psnr_v = psnr(gen_hr.squeeze().cpu().numpy(),imgs_hr.squeeze().cpu().numpy())
            ssim_v = ssim(gen_hr.squeeze().cpu().numpy(),imgs_hr.squeeze().cpu().numpy())
            
            
            writer.add_scalar('Loss/train_D_loss',loss_D.item(), i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_G_loss',loss_G.item(), i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_GAN_loss',loss_GAN.item(), i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_content_loss',loss_content.item(), i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_pixel_loss',loss_pixel.item(), i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_psnr',psnr_v, i+len(dataloader)*epoch)
            writer.add_scalar('Loss/train_ssim',ssim_v, i+len(dataloader)*epoch)
            
            print(f'[{i}/{len(dataloader)}],[{epoch}/{num_epoch}\t[D loss: {loss_D.item():.5f} G loss: {loss_G.item():.5f}]]')
            print('\n[SSIM: %.3f ]'%ssim_v)
            print('\n[PSNR: %.3f ]'%psnr_v)
            
            if (i+1)%save_every_iter == 0:    
                '''    
                torch.save({
                'epoch': epoch,
                'Gnet_state_dict': Gnet.state_dict(),
                'Dnet_state_dict': Dnet.state_dict(),
                'optimizer_G_state_dict': optimizer_G.state_dict(),
                'optimizer_D_state_dict': optimizer_D.state_dict()
                }, '../saved_models/%s_Crop_%d_%d.pth'%(model_name,epoch,i))
                '''
                plot_val(gen_hr.cpu(),imgs_hr.cpu(),"ESRGAN19_LN")

        # --------------Empty cache here-------------------------------
        writer.flush()
        torch.cuda.empty_cache()

writer.add_hparams(hparams,metrics)
writer.close()
```

```python
#%% training function
def train_loop(dataloader,model,loss_function,optimizer,device,epoch,num_epoch,checkpoint,
               :str="jupyter"):
    '''
    model = list
    optimizer = list
    '''
    warm_up = False # whether want a warm up session for Gnet
    epoch = epoch + checkpoint
    optimizer_D,optimizer_D = optimizer
    Dnet,Gnet,FE = model
    
    criterion_pixel,criterion_GAN,criterion_content=loss_function
    D_loss = []
    G_loss = []
    for i,imgs in enumerate(dataloader):
        
        imgs_hr = imgs['hr'].to(device)
        imgs_lr = imgs['lr'].to(device)
        valid = Variable(Tensor(np.ones((imgs_lr.size(0), *Dnet.output_shape))), requires_grad=False)
        fake = Variable(Tensor(np.zeros((imgs_lr.size(0), *Dnet.output_shape))), requires_grad=False)
        optimizer_G.zero_grad() # clean all grad at begining of update
        with autocast():
            gen_hr = Gnet(imgs_lr) 
            loss_pixel = criterion_pixel(gen_hr,imgs_hr)
        
        if (epoch==0 and warm_up == True):
            loss_pixel.backward()
            optimizer_G.step()
            print(
                "[Epoch %d/%d] [Batch %d/%d] [G pixel: %f]"
                % (epoch, num_epoch, i, len(dataloader), loss_pixel.item())
                 )   
            
            continue
        
        with autocast():
            pred_real = Dnet(imgs_hr).detach()
            pred_fake = Dnet(gen_hr)
            loss_GAN = criterion_GAN(pred_fake - pred_real.mean(0, keepdim=True), valid)
        
        
        with autocast():
            gen_features = FE(gen_hr) 
            real_features = FE(imgs_hr).detach()
        
            loss_content = criterion_content(gen_features, real_features)
            loss_G = loss_content + 5e-3 * loss_GAN + 1e-2 * loss_pixel
        
        scaler_G.scale(loss_G).backward()
        scaler_G.step(optimizer_G)
        scaler_G.update()
        #
        # Discriminator
        #
        optimizer_D.zero_grad()
        with autocast():
            pred_real = Dnet(imgs_hr)
            pred_fake = Dnet(gen_hr.detach())
    
        # Adversarial loss for real and fake images (relativistic average GAN)
            loss_real = criterion_GAN(pred_real - pred_fake.mean(0, keepdim=True), valid)
            loss_fake = criterion_GAN(pred_fake - pred_real.mean(0, keepdim=True), fake)
    
        # Total loss
            loss_D = (loss_real + loss_fake) / 2
    
        scaler_D.scale(loss_D).backward()
        scaler_D.step(optimizer_D)
        
        scaler_D.update()
        ## commented 12:19 am. 17 Juli, for ruling out checkboard 
        ## NOT influential...but training set did
        with torch.no_grad(): 
            D_loss.append(loss_D.item())
            G_loss.append(loss_G.item())
            print(f'[{i}/{len(dataloader)}],[{epoch}/{num_epoch}\t[D loss: {loss_D.item():.5f} G loss: {loss_G.item():.5f}]]')
            print('\n[SSIM: %.3f ]'%ssim(imgs_hr[0].squeeze().cpu().numpy(),gen_hr[0].squeeze().cpu().numpy()))
            print('\n[PSNR: %.3f ]'%psnr(y[0].squeeze().cpu().numpy(),imgs_hr[0].squeeze().cpu().numpy()))
            if i == (len(dataloader)-1):
            #if (i+1)%save_every_iter == 0:            
                torch.save(Gnet.state_dict(),'saved_models/Jupyter_Crop_generator_%d_%d.pth'%(epoch,i))
                torch.save(Dnet.state_dict(),'saved_models/Jupyter_Crop_discriminator_%d_%d.pth'%(epoch,i))

                os.makedirs('images/training/3dimg_crop_less_%s/'%(model_name),exist_ok=True)
                os.makedirs('saved_models/',exist_ok=True)                
                img_grid = torch.cat((imgs_hr[:,:,imgs_hr.size(2)//2], gen_hr[:,:,imgs_hr.size(2)//2]), -1).detach().cpu()
                save_image(img_grid, "images/training/3dimg_crop_less_%s/%d_%d.png" % (model_name,i,epoch), nrow=1, normalize=False)
            
            # torch.save(Dnet.state_dict(),'saved_models/C6_dicriminator_%d.pth'%epoch)
        # --------------Empty cache here-------------------------------
        torch.cuda.empty_cache()
        
        
        # Save image grid with upsampled inputs and ESRGAN outputs
        torch.cuda.empty_cache()
        del gen_hr
    return D_loss,G_loss

```

```python
D_loss = []
G_loss = []
for epoch in tqdm(range(num_epoch)):
    (D_l,G_l)=train_loop(data,
               model=[Dnet,Gnet,FE],
               loss_function = [criterion_pixel,criterion_GAN,criterion_content],
               optimizer= [optimizer_D,optimizer_G],
               device = device,epoch = epoch,num_epoch=num_epoch,checkpoint=checkpoint)
    D_loss.append(D_l)
    G_loss.append(G_l)
dict_save = {'D_loss':D_loss,
             'G_loss':G_loss}
os.makedirs('loss',exist_ok=True)
savemat(f'loss/table_3d_crop_ESRGAN_{num_epoch}ep.mat',dict_save,oned_as='row')

```

```python
def plot_func(a,b):
    img_a = a.detach().cpu().numpy().squeeze()
    img_b = b.detach().cpu().numpy().squeeze()
    fig = plt.figure()
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.imshow(img_a[int(img_a.shape[0]/2)])
    ax1.title.set_text("$\mu$=%.3f \sigma=%.3f,"%(img_a.mean(),img_a.std()))
    ax2.imshow(img_b[int(img_b.shape[0]/2)])
    ax2.title.set_text("$\mu$=%.3f \sigma=%.3f,"%(img_b.mean(),img_b.std()))
    plt.show()
```

```python
from skimage.metrics import structural_similarity
```

```python
def ssim(image,gt):
    data_range = np.max(gt) - np.min(gt)
    return structural_similarity(image, gt, data_range = data_range)

def psnr(image,gt):
    mse = np.mean((image - gt)**2)
    if mse == 0:
        return float('inf')
    data_range = np.max(gt) - np.min(gt)
    return 20* np.log10(data_range) - 10*np.log10(mse)
```
